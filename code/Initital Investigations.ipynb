```r
# import the csv as data frame
path <- "Games Database - Main Page.csv"
content <- read.csv(path)
```

```r
# have a look at the data
print (content)
```

```r
# tidy the data
cols.dont.want <- c("Game","Me...Review","Percentile") # choose columns to remove
df <- content[, ! names(content) %in% cols.dont.want, drop = F] # remove those columns

colnames(df)[3] <- "Launch Plaftorms"
colnames(df)[4] <- "Play Plaftorm"
colnames(df)[7] <- "Revisit Years"
colnames(df)[10] <- "DLC Played"
colnames(df)[11] <- "Remaster Played"
```

```r
# calculate correlation between rating and reviewscore
Corr <- cor(df$Rating, df$Reviewscore)
print("Correlation coefficient: ")
print(Corr)
# calculate linear fit 
fit <- lm(df$Rating ~ df$Reviewscore)
summary(fit)
```

```r
# create scatter plot of rating and reviewscore
plot(df$Rating, df$Reviewscore, xlab = "Review Average", ylab = "My Rating") #plot the scatter points
abline(fit, col = "orange", lwd = 3) #plot the line of best fit

text(paste("Correlation:", round(Corr, 2)), x = 40, y = 95) #plot the correlation coeff

cf <- round(coef(fit), 2) 
eq <- paste0("Score = ", cf[1],
             ifelse(sign(cf[2])==1, " + ", " - "), abs(cf[2]), " Review")
text(eq, x = 40, y = 90)
```
